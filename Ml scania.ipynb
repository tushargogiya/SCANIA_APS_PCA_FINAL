```python
l=[]
for index,i in enumerate(range(df_corr.shape[1])):
  if np.abs(df_corr.iat[0,i]) >= 0.3:
    l.append(index) 
print(l)
```

```python
lis=[df_corr.columns[i] for i in l ]
lis=lis[1:]
```

```python
df_highcorr_data=df_new[lis].copy()
```

```python
x=np.array(df_highcorr_data)
```

```python
cov_mat=np.cov(x,rowvar=False)
```

```python
cov_mat.shape
```

```python
eigen_value,eigen_vector=np.linalg.eigh(cov_mat)
```

```python
print(eigen_value.shape)
print(eigen_vector.shape)
```

```python
print(eigen_value)
print(eigen_vector)
```

```python
var=[]
for i in range(len(eigen_value)):
  var.append(eigen_value[i]/np.sum(eigen_value))
print(np.sum(var),var,"\n")

var.sort(reverse=True)

np.sum(var[0:20]) #30 columns contains 98 percent of variance so our n_component is equal to 30
```

```python
sorted_index = np.argsort(eigen_value)[::-1]
print(sorted_index)
sorted_eigenvalue = eigen_value[sorted_index]
sorted_eigenvector = eigen_vector[:,sorted_index]

sorted_eigenvalue, sorted_eigenvector
print(45*'/')
print(sorted_eigenvector.shape)
```
